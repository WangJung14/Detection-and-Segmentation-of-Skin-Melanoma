## 1. Load trọng số mô hình U-Net

```python
import torch
import cv2
import matplotlib.pyplot as plt
import sys
sys.path.append(r'd:\Computer Vision Final Project\Src code')
from src.u_net.inference import UNetInferencer
inferencer = UNetInferencer(weight_path=r'd:\Computer Vision Final Project\Src code\best_model.pth')
print("Da tai thanh cong best_model.pth vao U-Net!")
```

## 2. Inference (Dự đoán)

```python
img_path = r'd:\Computer Vision Final Project\Src code\data\train\images\ISIC_0000172.jpg'
gt_path = r'D:\Computer Vision Final Project\Src code\data\train\masks\ISIC_0000172_segmentation.png'
img = cv2.imread(img_path)
img = cv2.resize(img, (600, 450))
gt = cv2.resize(cv2.imread(gt_path, 0), (600, 450))
unet_mask = inferencer.predict(img)
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); axs[0].set_title('Anh goc'); axs[0].axis('off')
axs[1].imshow(gt, cmap='gray'); axs[1].set_title('Ground Truth'); axs[1].axis('off')
axs[2].imshow(unet_mask, cmap='gray'); axs[2].set_title('U-Net Du doan'); axs[2].axis('off')
plt.show()
```
